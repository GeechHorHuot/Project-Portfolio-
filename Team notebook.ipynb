```python
# target variable = covid_vaccination_vacc
```

```python
# imports for machine learning

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
# Putting feature variable to X
X = df_dummies.drop(["ID", "covid_vaccination_vacc"],axis=1)
# Putting response variable to y
y = df_dummies['covid_vaccination_vacc']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42, stratify=y)

rfc = RandomForestClassifier(random_state=42, class_weight= "balanced_subsample", max_depth=10, min_samples_leaf=2, min_samples_split=20, n_estimators=2000, max_features=5)
rfc.fit(X_test, y_test)
```

```python
y_pred = rfc.predict_proba(X_test)

from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_pred[:,1]))
print(roc_auc_score(y_test, y_pred[:,1]))
```

```python
y_pred = rfc.predict(X_test)
```

```python
y_pred_prob=rfc.predict_proba(X_test)[:,:]
y_pred_prob_df=pd.DataFrame(data=y_pred_prob, columns=['Prob of not vacinated (0)','Prob of vacinnated (1)'])
y_pred_prob_df.head(10)
```

```python
from sklearn.metrics import plot_roc_curve

plot_roc_curve(rfc, X_test, y_test)
plt.savefig("roc_curve.png")
```

```python
from sklearn.metrics import confusion_matrix

confusion = confusion_matrix(y_test, y_pred)
confusion
```

```python
conf_heatmap = sns.heatmap(confusion, annot=True, fmt="d", cmap="RdBu")
conf_heatmap.set_xlabel("Predicted")
conf_heatmap.set_ylabel("Actual")

plt.savefig("confusion matrix.png")
```

```python
imp_features = pd.DataFrame({
    "Varname": X_train.columns,
    "Imp": rfc.feature_importances_
})
```

```python
imp_features.sort_values(by="Imp", ascending=False, inplace=True)
imp_features_20 = imp_features.head(20)
```

```python
plt.figure(figsize = (20, 20))
plt.barh(imp_features_20["Varname"], imp_features_20["Imp"])
plt.title("The Importance of Features")
plt.gca().invert_yaxis()
plt.savefig("Imp_features.png")
```
